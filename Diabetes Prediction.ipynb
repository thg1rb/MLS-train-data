### ข้อมูลมีการ/น่าจะมีการสุ่มตัวอย่างมา (sampling) แบบใด
น่าจะมีการสุ่มตัวอย่างมาแบบ Simple Random Sample เนื่องจากข้อมูลของผู้ที่เป็นและไม่ได้เป็นเบาหวานมีความห่างกันมากจำนวนไม่ใกล้เคียงกัน 

```python
diabetes["Diabetes_binary"].value_counts()
```

### ข้อมูลมีการ/น่าจะมีการกำหนดเฉลย (labeling) ด้วยวิธีใด
